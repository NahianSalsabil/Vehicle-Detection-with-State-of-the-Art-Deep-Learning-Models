```
data = json.loads(uploaded[file_name].decode("utf-8"))
```

```
import pandas as pd
import numpy as np
# create dataframe
df = pd.json_normalize(data)
# save to csv
# df.to_csv('submission.csv', index=False, encoding='utf-8')
df
```

```
df.to_csv('something.csv', index=False)
```

```
new_df = df
new_df['image_id'] = new_df['filename']
for i in range(0,new_df['filename'].size) :
  new_df['image_id'].iloc[i] = new_df['filename'].iloc[i].split('/')[1]
new_df.drop(['filename','frame_id'],axis = 1)
```

```
value = []
image_id = []
count = 0
for i in range(0,new_df['objects'].size) :
  for j in range(0,len(new_df['objects'].iloc[i])):
    count = count + 1
    value.append(new_df['objects'].iloc[i][j])
    image_id.append(new_df['image_id'].iloc[i])
print(count)
```

```
column = pd.DataFrame({'col':np.arange(count)})
new_df = pd.concat([new_df,column], ignore_index=True, axis=1)
new_df.head(10)
```

```
new_df.drop([0,1,2,3],axis = 1,inplace= True)
# new_df
```

```
new_df.insert(0,"image_id",image_id)
new_df.insert(1, "column", value)
new_df.drop([4],axis = 1,inplace = True)
new_df
```

```

bbox_value = pd.DataFrame(new_df.column.values.tolist())['relative_coordinates']
bbox_value = pd.DataFrame.from_records(bbox_value)
bbox_value
```

```
#Formatting the Dataframe
class_name = []
score = []
for i in range(new_df['column'].size):
  class_name.append(new_df['column'].iloc[i]['name'])
  score.append(new_df['column'].iloc[i]['confidence'])
new_df.insert(1,"name",class_name)
new_df.insert(3, "score", score)
new_df = pd.concat([new_df,bbox_value],ignore_index=True, axis = 1)
new_df = new_df.rename(columns={0: "image_id", 1: "class", 3: "score", 4: "xmin", 5: "ymax", 6: "width", 7: "height"})
new_df.drop([2],axis = 1,inplace = True)
new_df['xmax'] = new_df['xmin'] + new_df['width']
new_df['ymin'] = new_df['ymax'] - new_df['height']
new_df.drop(['width','height'],axis = 1,inplace = True)
new_df['width'] = pd.Series([1024 for x in range(len(new_df.index))])
new_df['height'] = pd.Series([1024 for x in range(len(new_df.index))])
new_df = new_df[["image_id","class","score","xmin","ymin","xmax","ymax","width","height"]]
for i in range(new_df['xmax'].size):
  if new_df['xmax'].iloc[i] > 1024:
    new_df['xmax'].iloc[i] = 1024;
for i in range(new_df['ymax'].size):
  if new_df['ymax'].iloc[i] > 1024:
    new_df['ymax'].iloc[i] = 1024;
for i in range(new_df['class'].size):
  if new_df['class'].iloc[i] == "three_wheelers_(CNG)":
    new_df['class'].iloc[i] = "three wheelers (CNG)"
  if new_df['class'].iloc[i] == "human_hauler":
    new_df['class'].iloc[i] = "human hauler"
  if new_df['class'].iloc[i] == "auto_rickshaw":
    new_df['class'].iloc[i] = "auto rickshaw"
  if new_df['class'].iloc[i] == "army_vehicle":
    new_df['class'].iloc[i] = "army vehicle"
new_df
```

```
new_df.to_csv('submission.csv',index = False)
```
